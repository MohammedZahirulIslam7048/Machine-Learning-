# Library

```python
import numpy as np
import pandas as pd
```

# Dataset

```python
df = pd.read_csv('carprices.csv')
```

```python
df
```

```python
df.shape
```

```python
df.head(3)
```

# Preprocessing

```python
dummies = pd.get_dummies(df['Car Model'])
```

```python
dummies
```

```python
dummies1 = pd.get_dummies(df['Car Model_1'])
```

```python
dummies1
```

# Concat

```python
merged = pd.concat([df, dummies], axis = 'columns')
```

```python
merged
```

```python
merged.shape
```

```python
merged1 = pd.concat([df, dummies1], axis = 'columns')
```

```python
merged1
```

```python
merged1.shape
```

# Delete Column

```python
final = merged.drop(["Car Model"], axis = 1)
```

```python
final
```

```python
final1 = final.drop(["Car Model_1"], axis = 1)
```

```python
final1
```

# Separate X, Y

```python
x = final1.drop(['Sell Price'], axis = 1)
```

```python
x
```

```python
y = df['Sell Price']
```

# Algorithm (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
```

# Object

```python
model = LinearRegression()
```

# Fit Model

```python
model.fit(x, y)
```

# Score

```python
model.score(x,y)
```

# New Price Predict

```python
model.predict([[45000, 8, 1, 0, 0, 0]])
```
