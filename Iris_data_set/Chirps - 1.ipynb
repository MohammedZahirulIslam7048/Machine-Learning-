```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('chirps.csv')
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df[['Chirps_15s']]
y = df['Temp_C']
```

```python
x
```

```python
y
```

# Visualization

```python
#Scatter Plot
plt.scatter(df['Chirps_15s'], df['Temp_C'], marker = '+', color = 'blue')
plt.xlabel('Cricket Chirp / 15 Seconds')
plt.ylabel('Temperature / Celsius')
plt.title('Cricket Chirp and Temperature')
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

# Algorithm

```python
from sklearn.linear_model import LinearRegression
```

# Object

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain, ytrain)
```

```python
pred = reg.predict(xtest)
```

```python
pred
```

# R2_Score

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest, pred)
```

```python
score
```

# Plotting

```python
plt.scatter(df['Chirps_15s'], df['Temp_C'], marker = '+', color = 'orange')
plt.xlabel('Cricket Chirp / 15 Seconds')
plt.ylabel('Temperature / Celsius')
plt.title('Cricket Chirp and Temperature')
plt.plot(df.Chirps_15s, reg.predict(df[['Chirps_15s']]))
```
